```python
import pandas as pd
import json

DATASET_TYPE_LIST = ["NYC", "TKY", "CA"]
DATASET_SUBTYPE_LIST = ["train", "test"]

dataset_path = "../dataset"

for DATASET_TYPE in DATASET_TYPE_LIST:
    data_mapping_df = pd.read_csv(f"{dataset_path}/{DATASET_TYPE}/data_mapping_{DATASET_TYPE}.csv")

    for DATASET_SUBTYPE in DATASET_SUBTYPE_LIST:
        csv_df = pd.read_csv(f"{dataset_path}/{DATASET_TYPE}/split_data/{DATASET_SUBTYPE}_data.csv")

        join_df = pd.merge(csv_df, data_mapping_df, on=['Uid', 'Pid', 'Time'], how='inner')

        cat_groupby_df = join_df.groupby('Catname')['Pid'].agg(lambda x: x.unique().tolist()).reset_index()
        region_groupby_df = join_df.groupby('Region')['Pid'].agg(lambda x: x.unique().tolist()).reset_index()
        user_groupby_df = join_df.groupby('Uid')['Pid'].agg(lambda x: x.unique().tolist()).reset_index()
        
        cat_groupby_dict = cat_groupby_df.set_index('Catname')['Pid'].to_dict()
        region_groupby_dict = region_groupby_df.set_index('Region')['Pid'].to_dict()
        user_groupby_dict = user_groupby_df.set_index('Uid')['Pid'].to_dict()

        with open(f"{dataset_path}/{DATASET_TYPE}/graph/{DATASET_SUBTYPE}_catname_graph.json", "w") as f:
            json.dump(cat_groupby_dict, f, indent=4)
        
        with open(f"{dataset_path}/{DATASET_TYPE}/graph/{DATASET_SUBTYPE}_region_graph.json", "w") as f:
            json.dump(region_groupby_dict, f, indent=4)
        
        with open(f"{dataset_path}/{DATASET_TYPE}/graph/{DATASET_SUBTYPE}_user_graph.json", "w") as f:
            json.dump(user_groupby_dict, f, indent=4)
```
